## Let us split edition into the type of book and the publication date

#### We are doing the following to split the Edition into the publication type and its date. First, I am making sure there are no leading/tracing spaces:

## raw_data1['Edition'] = raw_data1['Edition'].str.strip()

#### Then, I am extracting the Publication type.  What the Regex does is to look for one or more non-digit characters. So once it hits the first digit it stops and thus gives me the name:

## raw_data1['Edition'].str.extract('(\D+)', expand=False)

#### To separate the the date/year of publication, I am using the same functionality. But here I am looking for the first digit to appear and any following character.

## raw_data1['Edition'].str.extract('(\d+.*)', expand=False)</p>

```python
raw_data1['Edition'] = raw_data1['Edition'].str.strip()
raw_data1['Publication_Type']=raw_data1['Edition'].str.extract('(\D+)', expand=False)
raw_data1['Publication_Date'] = raw_data1['Edition'].str.extract('(\d+.*)', expand=False)
raw_data1['Publication_Type']=raw_data1['Publication_Type'].str.replace('\,–',"")
raw_data1['Publication_Type']=raw_data1['Publication_Type'].str.replace('\,',"")
```

```python
raw_data['Edition'] = raw_data['Edition'].str.strip()
raw_data['Publication_Type']=raw_data['Edition'].str.extract('(\D+)', expand=False)
raw_data['Publication_Date'] = raw_data['Edition'].str.extract('(\d+.*)', expand=False)
raw_data['Publication_Type']=raw_data['Publication_Type'].str.replace('\,–',"")
raw_data['Publication_Type']=raw_data['Publication_Type'].str.replace('\,',"")
```

```python
test_data['Edition'] = test_data['Edition'].str.strip()
test_data['Publication_Type']=test_data['Edition'].str.extract('(\D+)', expand=False)
test_data['Publication_Date'] = test_data['Edition'].str.extract('(\d+.*)', expand=False)
test_data['Publication_Type']=test_data['Publication_Type'].str.replace('\,–',"")
test_data['Publication_Type']=test_data['Publication_Type'].str.replace('\,',"")
```

```python
raw_data['Publication_Type']=raw_data['Publication_Type'].str.replace('Jan','').str.strip()
raw_data['Publication_Type']=raw_data['Publication_Type'].str.replace('Feb','').str.strip()
raw_data['Publication_Type']=raw_data['Publication_Type'].str.replace('Mar','').str.strip()
raw_data['Publication_Type']=raw_data['Publication_Type'].str.replace('Apr','').str.strip()
raw_data['Publication_Type']=raw_data['Publication_Type'].str.replace('May','').str.strip()
raw_data['Publication_Type']=raw_data['Publication_Type'].str.replace('Jun','').str.strip()
raw_data['Publication_Type']=raw_data['Publication_Type'].str.replace('Jul','').str.strip()
raw_data['Publication_Type']=raw_data['Publication_Type'].str.replace('Aug','').str.strip()
raw_data['Publication_Type']=raw_data['Publication_Type'].str.replace('Sep','').str.strip()
raw_data['Publication_Type']=raw_data['Publication_Type'].str.replace('Oct','').str.strip()
raw_data['Publication_Type']=raw_data['Publication_Type'].str.replace('Nov','').str.strip()
raw_data['Publication_Type']=raw_data['Publication_Type'].str.replace('Dec','').str.strip()
```

```python
raw_data['Publication_Type'].value_counts()
```

```python
def mapping(df,col,n=25):
 print(col,n)
 vc = df[col].value_counts()
 replacements = {}
 for col, s in vc.items():
    print(col,s)
    if s<n:
        replacements[col] = 'other'
 return replacements
```

```python
Pub_type_replace = mapping(raw_data,'Publication_Type',9)
```

```python
raw_data['Publication_Type'] = raw_data['Publication_Type'].replace(Pub_type_replace)
```

```python
raw_data['Publication_Type'].value_counts()
```

```python
raw_data['Publication_Date'] = pd.to_datetime(raw_data['Publication_Date'])
raw_data['Publication_Year'] = pd.DatetimeIndex(raw_data['Publication_Date']).year
raw_data['Publication_Month'] = pd.DatetimeIndex(raw_data['Publication_Date']).month
raw_data.drop(['Edition','Publication_Date'], axis=1, inplace=True)
```

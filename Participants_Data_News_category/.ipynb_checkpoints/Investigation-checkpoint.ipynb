```python
def display_all(df):
    with pd.option_context("display.max_rows", 1000, "display.max_columns", 1000): 
        display(df)
```

```python
from nltk.stem import WordNetLemmatizer  # lammatizer from WordNet
from nltk.corpus import stopwords 
import re
#Clean the text by retaining only alphabets and converting all the characters to small
def clean_up(review):
    clean = re.sub("[^a-zA-Z]"," ",review)
    #replace multiple space by a single
    clean = re.sub(' +', ' ',clean)
    
    word_tokens= clean.lower().split()
    
    # 4. Remove stopwords
    le=WordNetLemmatizer()
    stop_words= set(stopwords.words("english"))     
    word_tokens= [le.lemmatize(w) for w in word_tokens if not w in stop_words]
    
    cleaned =" ".join(word_tokens)
    #re.sub(' +', ' ',string4)
    return cleaned
```

```python
raw_data1['cleaned'] = raw_data1['STORY'].apply(clean_up)
```

```python
display_all(raw_data1)
```

```python
raw_data1['Text_Length'] = raw_data1['cleaned'].str.len()
```

```python
#Average count of words for different words
import math
for i in range(0,4):
    print('Section',"",i,"",math.ceil(raw_data1[raw_data1['SECTION']==i]['Text_Length'].mean()))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.preprocessing import text, sequence
from keras import utils
```

```python
train_snippets = raw_data1['cleaned']
snippet_y_train = utils.to_categorical(raw_data1['SECTION'], 4)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.preprocessing import text, sequence
from keras import utils
```

```python
#Create vectors of the text
max_words = 1000
tokenize = text.Tokenizer(num_words=max_words, char_level=False)
tokenize.fit_on_texts(train_snippets) # only fit on train
snippet_x_train = tokenize.texts_to_matrix(train_snippets)
```

```python
snippet_y_train.shape
```

```python
batch_size = 32
epochs = 10

# Build the model 1
model = Sequential()
model.add(Dense(512, input_shape=(max_words,)))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(4))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
              
history = model.fit(snippet_x_train, snippet_y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_split=0.2)
```

```python
test_data['cleaned'] = test_data['STORY'].apply(clean_up)
```

```python
# Make vectors of the text from the test data
test_snippets = test_data['cleaned']
snippet_x_test = tokenize.texts_to_matrix(test_snippets)
```

```python
predictions_are = model.predict_classes(snippet_x_test)
```
